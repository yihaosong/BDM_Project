```python
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
with open('nyc_request_type_info') as fi1:
    count = fi1.read()
    count = count.replace(')', '')
    count = count.replace('(', '')
out1 = open('nyc_request_type_info.txt','w+')
out1.write(count)
out1.close()
data = pd.read_csv('nyc_request_type_info.txt', sep='\t', header = None)
print len(data)
print data

dict = {}
for d in data[0]:
    dict[d.split(',')[0].split("'")[1]] = int(d.split(',')[1])
dict_key = sorted(dict)

service_list1 = []
count_list1 = []
for d in dict_key:
    service_list1.append(d)
    count_list1.append(dict[d])

max_count = max(count_list1)
print max_count
for i in range(len(service_list1)):
    if dict[service_list1[i]] == max_count:
        print service_list1[i]

plt.figure(figsize=(20,10))
plt.title('NYC 311 Request Type counts',fontsize=18)
plt.xticks(range(len(count_list1)), service_list1, rotation='vertical',fontsize=18)
plt.ylabel('Number of calls', fontsize=18)
plt.bar(range(len(count_list1)),count_list1)
plt.axes().grid(True)
```
